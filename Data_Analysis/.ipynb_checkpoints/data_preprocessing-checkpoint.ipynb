## Importing Libraries


```python
# Dependencies
import pandas as pd
from pathlib import Path
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

# Loading the dataset

```python
# Read the dataset file with the Pandas 
dataset_df = pd.read_csv("../Resources/WA_Fn-UseC_-HR-Employee-Attrition.csv")

# Display the dataframe
dataset_df.head()
```

## Data Wrangling

```python
# Dimension of dataset
dataset_shape = dataset_df.shape
print(f"Number of rows in dataset: {dataset_shape[0]}")
print(f"Number of columns in dataset: {dataset_shape[1]}")
```

```python
# Dataset infomation
dataset_type = dataset_df.info()
dataset_type
```

```python
# Count the numerical and categorical columns
numerical_count = 0
categorical_count = 0

for column_name, data_type in dataset_df.dtypes.items():
    if np.issubdtype(data_type, np.number):
        numerical_count += 1
    else:
        categorical_count += 1

# Print the counts
print(f"There are {numerical_count} Numerical Columns in dataset")
print(f"There are {categorical_count} Categorical Columns in dataset")
```

```python
# Statistical Summary of Dataset
dataset_df.describe()
```

```python
# Non-statistical Summary of Dataset
dataset_df.describe(include= 'object')
```

```python
# Checking if there is any duplicate records.
print(f"Duplicates in Dataset: {dataset_df.duplicated().sum()}")
```

```python
# Checking if there is any Null records.
print(f"Null values in Dataset: {dataset_df.isnull().sum().sum()}")
```

```python
# Identify incomplete rows
dataset_df.count()
```

```python
# Replace underscores with spaces in the symptoms
for col in dataset_df.columns:
    if dataset_df[col].dtype == 'object':
        dataset_df[col] = dataset_df[col].str.replace('_',' ')
```

```python
# Print all columns in dataset
dataset_columns = dataset_df.columns
dataset_columns
```

```python
# Dropping unnecessary columns which does not imply any meaningful insights in the analysis
unnecessary_columns = ["Over18", "EmployeeCount", "EmployeeNumber", "StandardHours"]
final_df = dataset_df.drop(columns= unnecessary_columns)

# Display the dataframe
final_df.head()
```
